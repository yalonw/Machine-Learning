## [numpy.random.normal](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html) 
- numpy.random.normal(loc=0.0, scale=1.0, size=None) 
- 高斯常態分佈：中心點 loc=0，標準差 scale=1
- 標準差和平均數，大多用在常態分布

```python
import numpy as np

dist = np.random.normal(loc=10, scale=1,  size=100000)
dist = dist[dist>=0]
dist
```

## [scipy.stats.boxcox_normmax](https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.boxcox_normmax.html)


```python
from scipy.stats import boxcox_normmax
```

## [sns.distplot()](https://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# 高斯常態分布
print(boxcox_normmax(dist))
sns.distplot(dist) # kde=True (must)
```

## [numpy.power](https://docs.scipy.org/doc/numpy/reference/generated/numpy.power.html)

```python
# 平方：右偏 (差一點就差很多)
# 次方型右偏，需要透過 反向次方 來修正為常態分佈
dist1 = np.power(dist, 6)
print(boxcox_normmax(dist1))
sns.distplot(dist1)
```

```python
dist1 = dist1**(0.17163888690958568)
sns.distplot(dist1)
```

```python
# 開根號：左偏 (影響不大)
dist2 = np.power(dist, 1/10000)
print(boxcox_normmax(dist2))
sns.distplot(dist2)
```

```python
# 指數成長：極度右偏
# 指數型右偏，需要透過 log 來修正為常態分佈
dist3 = np.power(3,dist)
print(boxcox_normmax(dist3))
sns.distplot(dist3)
```
